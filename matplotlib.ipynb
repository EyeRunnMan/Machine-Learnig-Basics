```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
```

```python
plt.plot([1,3,5],[1,5,4],marker="*")
plt.show()
```

```python
x_squared=x**2
```

```python
plt.plot(x,x,marker="*")
plt.show()
```

```python
plt.plot(x,x_squared,marker="*")
plt.plot(x,x,marker="o")
plt.show()
```

```python
plt.plot(x,x_squared,marker="*")
plt.show()
plt.plot(x,x,marker="o")
plt.show()
```

```python
plt.plot(x,x_squared,marker="*")
plt.plot(x,x,marker="o")
plt.xlabel("Numbers")
plt.ylabel("Compensation")
plt.title("First Graph")
plt.show()
```
